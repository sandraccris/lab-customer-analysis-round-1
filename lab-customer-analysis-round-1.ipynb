```python
#CONCATENATE ALL 3 DATAFRAMES

data = pd.concat([df1, df2, df3])
data.head()
```

```python
data.columns
```

```python
#STANDARDIZING COLUMNS NAMES
cols = [col_name.lower().replace(' ', '_') for col_name in data.columns]
cols
```

```python
#APPLYING CHANGES TO DATAFRAME

data.columns = cols
data
```

```python
#ONLY NUMERICAL VALUES

data.describe()
```

```python
#WHICH ONES ARE CATEGORICAL? OBJECT ONES ARE CATEGORICAL

data.info()
```

```python
data.describe
```

```python
#DELETE COLUMN EDUCATION AND NUMBER OF OPEN COMPLAINS COLUMN

data2 = data.drop (columns = ["education", "number_of_open_complaints"])
data2.columns
```

```python
#REMOVING "%" FROM CUSTOMER_LIFETIME_COLUMN

data2["customer_lifetime_value"] = data2["customer_lifetime_value"].str.replace(r"%", "")
```

```python
data2
```

```python
#CHECKING NULL VALUES IN CUSTOMER_LIFETIME_VALUE COLUMN

data2["customer_lifetime_value"].isna().value_counts()
```

```python
#REPLACE NULLS WITH "0" IN CUSTOMER_LIFETIME_VALUE_FIXED COLUMN

data2["customer_lifetime_value"]=data2["customer_lifetime_value"].fillna("0")
```

```python
#CONVERTING TO FLOAT

data2["customer_lifetime_value"]=data2["customer_lifetime_value"].astype(float)
```

```python
#ROUNDING SO THERE ARE NO NUMBERS AFTER DECIMAL POINT, ROUNDING TO THE NEAREST 1
data2["customer_lifetime_value"]=round(data2["customer_lifetime_value"],0)
```

```python
#CONVERTING TO INTEGER

data2["customer_lifetime_value"]=data2["customer_lifetime_value"].astype(int)
```

```python
data2["customer_lifetime_value"]=data2["customer_lifetime_value"]*100
```

```python
data2["customer_lifetime_value"]
```

```python
#CHECK FOR DUPLICATES AND REMOVE THEM IF ANY

data3=data2.drop_duplicates(subset=["customer"], keep=False)
```

```python
data3
```

```python
##Filter out the data for customers who have an income of 0 or less.

data4=data3.query("income > 0")
```
